# Project Scope: Life Expectancy and GDP

## Goals
- Analyze the relationship between **GDP** and **Life Expectancy** for six countries.
- Answer key questions:
  - Is there a positive correlation between GDP and life expectancy?
  - How has life expectancy changed over time in these countries?
  - Are there exceptions (countries with high GDP but relatively low life expectancy, or the opposite)?

## Data
- Source: World Health Organization (WHO) and World Bank (provided in `all_data.csv`).
- Main columns: `Country`, `Year`, `GDP`, `Life Expectancy`.
- Preparation steps:
  - Check for missing values.
  - Ensure numerical columns (GDP, Year, Life Expectancy) are properly formatted.

## Analysis
- Descriptive statistics (mean, min, max, distributions).
- Visualizations:
  - Scatter plot: GDP vs. Life Expectancy.
  - Line plots: Life Expectancy over time by country.
- Insights:
  - Identify overall trends and correlations.
  - Highlight differences and anomalies between countries.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("all_data.csv")

# Take a look at the first few rows 
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.rename(columns={
    "Life expectancy at birth (years)": "Life_Expectancy"
})
```

```python
# Summary stats grouped by country
print(df.groupby("Country")["Life_Expectancy"].mean())
print(df.groupby("Country")["GDP"].mean())
```

### Insights
- Most countries have life expectancy between 74â€“80 years, except Zimbabwe which is significantly lower.
- GDP levels vary widely, with the USA and China far ahead.
- There seems to be a general correlation between higher GDP and higher life expectancy, though not always proportional (e.g., USA vs. Chile).

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.lineplot(x="Year", y="Life_Expectancy", hue="Country", data=df, marker="o")
plt.title("Life Expectancy over Time (2000-2015)")
plt.show()
```

**Life Expectancy over Time (2000â€“2015)**  
   Most countries show a steady increase in life expectancy, while Zimbabwe experienced a sharp rise after 2005.

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="Year", y="GDP", hue="Country", data=df, marker="o")
plt.title("GDP over Time (2000-2015)")
plt.show()
```

**GDP over Time (2000â€“2015)**  
   China and the United States show significant GDP growth, while other countries have slower, smaller growth trends.

```python
plt.figure(figsize=(8,6))
sns.lineplot(x="GDP", y="Life_Expectancy", hue="Country", data=df)
plt.title("GDP vs Life Expectancy")
plt.show()
```

**GDP vs Life Expectancy**  
   There is a positive correlation: countries with higher GDP tend to have higher life expectancy, with Zimbabwe as an outlier.


```python
plt.figure(figsize=(8,6))
sns.histplot(df["Life_Expectancy"], bins=15, kde=True)
plt.title("Distribution of Life Expectancy")
plt.show()
```

**Distribution of Life Expectancy**  
   Most values are concentrated between 75â€“80 years, with lower values mainly from Zimbabwe.
